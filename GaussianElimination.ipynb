```python
import numpy as np
import re
```

```python
def is_all_cap(string_list, m, mat):
    p_flag = False
    m_flag = False
    e_flag = False
#     o_flag = False
    s_flag = False
    for i in range(m):
        string = string_list[i].split()
#         print(string)
        s_flag = True
        k=0
#         val = 1
        for j in string:
#             print(j)
            if s_flag or e_flag:
                if j[0]=='-':
                    m_flag = True
                    j=j.replace('-','')
#                     print(j)
                else:
                    p_flag = True
#                 if 64<j[0] and j[0]<91:
#                     mat[i][k++] 
#             if len(j) == 1: and !e_flag and !p_flag and !m_flag:
            if j == "+":
                p_flag = True
            elif j == "-":
                m_flag = True
            elif j == "=":
                e_flag = True
            else:
                if not e_flag:
                    x = re.search("[0-9]*[A-Z]$",j)
#                     print(x)
                    if(x):
                        val = re.findall("[0-9]+",j)
#                         print(val)
                        if val:
                            if p_flag:
                                mat[i][k] = int(val[0])
                            elif m_flag:
                                mat[i][k] = -1*int(val[0])
                            k +=1
                        else:
                            if m_flag:
                                mat[i][k] = -1
                            elif p_flag:
                                mat[i][k] = 1
                            k +=1
                    else:
                        return mat,False
                else:
                    x = re.search("[0-9]*$",j)
#                     print(x)
                    if(x):
                        val = re.findall("[0-9]+",j)
#                         print(val)
                        if val:
                            if p_flag:
                                mat[i][k] = int(val[0])
                            elif m_flag:
                                mat[i][k] = -1*int(val[0])
                            k +=1
                        else:
                            if m_flag:
                                mat[i][k] = -1
                            elif p_flag:
                                mat[i][k] = 1
                            k +=1
                    else:
                        return mat,False
                p_flag = False
                m_flag = False
                e_flag = False
#                 o_flag = False
#                 s_flag = False
            s_flag=False
    return mat,True
                        
            
```
