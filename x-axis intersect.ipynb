```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.zeros(20)
xAxes = np.zeros(20)
xAxesY = np.zeros(20)
yAxes = np.zeros(20)
yAxesX = np.zeros(20)
i= -13
j=0
while i<=7 and j<20:
    X[j] = i
    xAxesY[j]=j-10
    yAxesX[j]=j-10
    i += 1
    j +=1
Y = X+3
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(xAxes,xAxesY)
plt.plot(yAxesX,yAxes)
plt.plot(X,Y)
plt.scatter([-3],[0])
plt.show()
```
